# K-Nearest Neighbors

## 1. visualization

```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
%matplotlib inline
```

```python
# load Iris datasets 
iris = datasets.load_iris() 

iris.data
```

```python
# here we only select sepal length and width (select first 2 columns)
X = iris.data[:, :2] #sepal len, sepal width, petal len, petal width (꽃받침, 꽃잎)
y = iris.target 
```

```python
X #sepal length, sepal width (꽃받침)
```

```python
y # 0: 'setosa', 1: 'versicolor', 2: 'virginica'
```

```python
X
```

```python
# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```

## 2. Creating a K-Nearest Neighbor Classifier

```python
# Load libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Create standardizer(정규화)
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
standardizer = StandardScaler()
```

```python
# Standardize features
X_std = standardizer.fit_transform(X)
```

```python
X_std
```

```python
# Plot the training points
plt.scatter(X_std[:, 0], X_std[:, 1], c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```

```python
# Train a KNN classifier with 3 neighbors
knn = KNeighborsClassifier(n_neighbors=3).fit(X_std, y)
```

```python
new_obs = [[ 0.0, 0.0],
            [-2.0,1.0],
           [2.0, -2.0]]
```

```python
knn.predict(new_obs)
```
